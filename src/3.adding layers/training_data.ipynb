```python
from nnfs.datasets import spiral_data
```

```python
import numpy as np
import nnfs

nnfs.init()
```

nnfs.init() does the following:
* sets random seed to 0
* creates a float32 datatype defaul
* overrides the original dot product from numpy  
This is to ensure repeatability in results

The sprial_data function allows to create a dataset with as many classes as we would want. The function has parameters to choose number of classes and number of observations per class in the resulting non-linear dataset.

```python
import matplotlib.pyplot as plt
```

```python
X, y = spiral_data(samples=100, classes=3)

plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("col 1")
plt.ylabel("col 2")
plt.title("scatterplot of spiral data")
plt.show()
```

Adding color to the plot makes the three classes easier to see.

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='brg')
plt.show()
```

Bear in min that the neural network will not be aware of the color differences as the data have no class encoding. This is only made for us as reader. Each dot is a feature and the coordinates are the samples that form the dataset. The classification for a dot depends on which spiral it is part of, blue, green or red in the plot. The colors would then be assigned a classification number to fit to, 0, 1, 2.
